### Data Selection Justification for Inflation Prediction

For the purpose of predicting inflation in Australia, we have selected the following columns from the dataset:

1. **CPI Index Numbers (Australia)**: The 'Index Numbers ; All groups CPI ; Australia' column represents the Consumer Price Index for all categories across the entire nation. This index is the most comprehensive representation of inflation, as it tracks changes in the price level of a basket of consumer goods and services purchased by households.

2. **Year-over-Year Percentage Change (Australia)**: The 'Percentage Change from Corresponding Quarter of Previous Year ; All groups CPI ; Australia' column is crucial for understanding the annual inflation rate. It measures the percentage change in the CPI compared to the same quarter of the previous year, offering insights into long-term inflation trends without the noise of seasonal effects.

3. **Quarter-over-Quarter Percentage Change (Australia)**: The 'Percentage Change from Previous Period ; All groups CPI ; Australia' column provides data on the inflation rate from one period to the next, which is essential for short-term analysis and can indicate the immediate direction of inflation trends.

These columns were chosen because they collectively provide a clear picture of both current price levels and their changes over time, which are the fundamental components needed for accurate inflation forecasting. The CPI Index Numbers give us the absolute price levels, while the percentage change columns show relative changes, allowing the model to account for both magnitude and momentum of inflation.

In our forecasting model, we will be utilizing these columns to train on historical data and predict future inflation rates. The model's performance will be evaluated based on how well it captures the trend and volatility of historical inflation rates and its accuracy in out-of-sample predictions.
